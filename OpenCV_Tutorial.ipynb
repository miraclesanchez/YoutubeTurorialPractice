```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/green_bird.jpg")

#when you read an image in openCV using imshow, every pixel in the image will be in BGR
resized_img = cv2.resize(img,(500,500))
cv2_imshow(resized_img)
print (resized_img.shape)


#this converts to the RGB color space
newImg = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)
cv2_imshow(newImg)
print (newImg.shape)
```
